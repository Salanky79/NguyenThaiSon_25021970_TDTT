```python
def is_perfect(n):
    divisors = []
    
    for i in range(1, n):
        if n % i == 0:
            divisors.append(i)

    total = sum(divisors)

    if total == n:
        print("Day la so perfect")
    else:
        print("Day ko phai la so perfect")
```

```python
def check_list(nums, k):
    for i in range(len(nums)):
        if nums[i] == k:
            print(f"So dau dien trong list xuat hien giong so k la o vi tri: {i}")
            return i
    return -1
        
```

```python
def factorial(n):
    i = n
    total = 1
    while i >= 1:
        total = total * i 
        i -= 1
    return total
```

```python
def calculator(num1, num2, operat):

    ket_qua = 0.0
    
    if operat == '+':
        ket_qua = num1 + num2
    elif operat == '-':
        ket_qua = num1 - num2
    elif operat == '*':
        ket_qua = num1 * num2
    elif operat == '/':
        
        if num2 == 0:
            
            print("Lỗi: Không thể chia cho 0.")
            return None 
        ket_qua = num1 / num2
    else:
        
        print(f"Lỗi: Toán tử '{operat}' không hợp lệ.")
        return None

    return ket_qua
```

```python
def check_trans(a, b):
    if len(a) != len(b):
        return False

    lst1 = [-1] * 26  # map a → b
    lst2 = [-1] * 26  # map b → a

    for i in range(len(a)):
        index_a = ord(a[i]) - ord('a')
        index_b = ord(b[i]) - ord('a')

        # Nếu đã có ánh xạ a → b nhưng khác b[i] → sai
        if lst1[index_a] != -1 and lst1[index_a] != b[i]:
            return False

        # Nếu đã có ánh xạ b → a nhưng khác a[i] → sai
        if lst2[index_b] != -1 and lst2[index_b] != a[i]:
            return False

        lst1[index_a] = b[i]
        lst2[index_b] = a[i]

    return True
```
