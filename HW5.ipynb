```python
#W6A19
def check_min(lst):
    min_idx = 0
    n = len(lst)
    for i in range(1, n):
        if lst[i] < lst[min_idx]:
            min_idx = i
    return min_idx + 1 
```

```python
def check(lst, k):
    n = len(lst)
    for i in range(n):
        if lst[i] == k:
            return i + 1
    return -1
```

```python
lst = input("Nhap vao cac chuoi: ").split()

def linear(lst):
    dct = {}
    for i in range(len(lst)):
        if lst[i] not in dct:
            dct[lst[i]] = [i]
        else:
            dct[lst[i]].append(i)

    for key in dct:
        dct[key] = tuple(dct[key])

    return dct

```

```python
def delete(lst, x):
    n = len(lst)
    i = 0
    while i < n:
        if lst[i] < x:
            lst[i] = lst[n - 1]
            n -= 1
        else:
            i += 1
    return lst[:n]
```

```python
def check_zero(lst):
    n = len(lst) - 1
    i = 0
    while i <= n:
        if lst[i] == 0:
            lst[i], lst[n] = lst[n], lst[i]
            n -= 1
        else:
            i += 1
    return lst
```

```python
def check_missing(lst):
    arr = set(lst)
    res = []
    for i in range(10):
        if i not in arr:
            res.append(i)
    return res
```

```python
def sum_local_max(lst):
    total = 0
    for i in range(1, len(lst) - 1):
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            total += lst[i]
    return total
```

```python
def check_x(lst, x):
    n = len(lst) - 1
    i = 0
    res = []
    while i <= n:
        if lst[i] < x:
            res.append(i + 1)
        i += 1
    return res
            
```

```python
def check_lst(lst):
    if len(lst) < 2:
        return True

    increasing = decreasing = True

    for i in range(len(lst) - 1):
        if lst[i] < lst[i+1]:
            decreasing = False
        elif lst[i] > lst[i+1]:
            increasing = False

    return increasing or decreasing
```
