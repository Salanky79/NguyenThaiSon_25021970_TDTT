```python
input_line = input("Nhập các cặp key:value (cách nhau bởi dấu cách, ví dụ: a:1 b:1 a:2): ")

dict = {}
for pair in input_line.split():
    if ':' in pair:
        key, value = pair.split(":", 1)

        if key in dict:
            dict[key].append(value)
        else:
            dict[key] = [value]

print(dict)
```

```python
def dem_tan_suat_chuoi_thu_cong(danh_sach_chuoi):
    tan_suat = {}
    for chuoi in danh_sach_chuoi:
        if chuoi in tan_suat:
            tan_suat[chuoi] += 1
        else:
            tan_suat[chuoi] = 1
    return tan_suat
```

```python

def add_dict(dict1, dict2):
    result = {}
    for key, value in dict1.items():
        result[key] = value
    for key, value in dict2.items():
        current_value = result.get(key, 0)
        result[key] = current_value + value

    sorted_result = dict(sorted(result.items()))
    return sorted_result
```

```python
def check(lst, k):
    seen = set()
    res = set()
    for x in lst:
        y = k - x
        if y in seen:
            res.add(tuple(sorted((x, y))))
        seen.add(x)
    return sorted(res)

```

```python
def tach_chan_le_tu_tuple(input_tuple):

    so_chan_list = []
    so_le_list = []
    

    for so in input_tuple:
        if so % 2 == 0:
            so_chan_list.append(so)
        else:
            so_le_list.append(so)
            
    tuple_chan = tuple(so_chan_list)
    tuple_le = tuple(so_le_list)
    
    return tuple_chan, tuple_le

```

```python
def check_list(nums):
    dict_check = {}
    for num in nums:
        dict_check[num] = dict_check.get(num, 0) + 1

    if not dict_check:
        return None
    max_check = max(dict_check.values())

    max_frequent = []
    for key, value in dict_check.items():
        if value == max_check:
            max_frequent.append(key)
    
    if len(max_frequent) > 1:
        return min(max_frequent)
    elif len(max_frequent) == 1:
        return max_frequent[0] if max_frequent else None
        
```

```python
def reverse_dict(ori_dict):
    new_dict = {}
    for key, value in ori_dict.items():
        new_dict[value] = key
    return new_dict
```

```python
def dict_filter(ori_dict, k):
    new_dict = {}
    for key, value in ori_dict.items():
        if value > k:
            new_dict[key] = value
    
    return new_dict
```

```python
def check(lst):
    stack = []
    pairs = {')': '(', ']': '[', '}': '{'}
    for char in lst:
        if char in '([{':
            stack.append(char)
        elif char in ')}]':
            if not lst or lst[-1] != pairs[char]:
                return False
            stack.pop()
    return len(stack) == 0

```
